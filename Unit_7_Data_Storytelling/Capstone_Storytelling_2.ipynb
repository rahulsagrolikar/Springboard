```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df_solo1000 = pd.read_csv('df_solo1000.csv') #In a previous notebook, I divided the datasets into solo, duo, and squad mode, and downlaoded them as csv files.
df_duo1000 = pd.read_csv('df_duo1000.csv')
df_squad1000 = pd.read_csv('df_squad1000.csv')
```

```python
first = df_squad1000.player_survive_time > 35
```

```python
print(df_squad1000[first].count()) #1. Only 10 players out of a 1000 survived over 35 minutes in a match. Thats 1%
```

```python
kills_mean = df_solo1000.groupby('team_placement')['player_kills'].mean().head(20) #2. Grouping by team placement, and summing 
# the total damage of players, we can observe a positive correlation between higher team placement and damage done
print(kills_mean)
```

```python
top40solo = df_solo1000[(df_solo1000.team_placement) < 40]
top40duo = df_duo1000[(df_duo1000.team_placement) < 40]
top40squad = df_squad1000[(df_squad1000.team_placement) < 40]
```

```python
fig = plt.figure(figsize = (12, 6)) #3

plt.subplot(1, 3, 1) #3

plt.title('Solo')
plt.bar(x='team_placement', height='player_kills', data=top40solo) #3

plt.subplot(1, 3, 2)
plt.title('Duo')

plt.bar(x='team_placement', height='player_kills', data=top40duo)

plt.subplot(1, 3, 3)
plt.title('Squad')

plt.bar(x='team_placement', height='player_kills', data=top40squad)
```

```python
sns.jointplot(x='player_dist_walk', y='player_kills', data=df_solo1000, kind='hex') #4
```

```python


sns.pairplot(df_solo1000[['player_kills', 'player_dist_walk']], size=3) #4 Most of the killings happen within the first 1000 feet of player movement
plt.title('Solo', loc='left')

sns.pairplot(df_duo1000[['player_kills', 'player_dist_walk']], size=3)
plt.title('Duo', loc='left')

sns.pairplot(df_squad1000[['player_kills', 'player_dist_walk']], size=3)
plt.title('Squad', loc='left')
```
