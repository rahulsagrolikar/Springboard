## Cross validation and metric


**Chosen cross-validation scheme**: A simple 80/20 split. I first group data into 5 folds. Then fold_0 is assigned to `test`, which will only be tested once for each method. The remaining data is assigned to `train`. `train` is further broken down into `trn` and `tst`. For each method we fit to `trn` and predict `tst` and `test`, giving cv_score and test_score. Each method may or may not be tuned using the cv_score, but never the . Both the cv_score and test_score could be 

### Sacrificing robustness for 96% reduction in computational expense and time.
To assess each method, we could run nested validation, with outer nesting being the "test" set, inner nesting being the "validation" set. For 5-fold cv, this would mean building 25 models. For the computational cost, we would gain proper estimation of our models performance and robustness. Instead, for each method, using one outer "test" split, and one inner "validation" split, building one model to get both "validation" and "test" score.

   * What would 5 fold outer CV achieve? The best test of robustness would be to use all 5 `train`/`test` splits available, completing all training steps for each iteration, including eda, feature selection, and hyper-parameter tuning, leading to robust choices for all steps of building the model. 
 2) Same as 1, but for the `trn`/`tst` splits. Here we could also get more robust models and hyper-parameters if we tune them using k-fold validation. 
 3) Not even retraining each model on all `train` data, which should yield better models. We are using the model trained with `trn` to predict `test`, even though we could have 33% more data using `train`.

**Chosen metric**: Accuracy - Since this is a ternary-classification problem with balanced labels, accuracy should be a reasonable metric. If we decide that we care more about one class than another, or if we want to first predict if a user is a brand or a person first, then predict gender, we could adjust our metric and modeling methods.

**Considerations**: 

 - **Time consideration**: Since there is no date-time for the tweet, we will make no consideration for time, but in general we can only build models on data from the past, so this should be included in future cross-validation schemes. 

 - **`names` with multiple tweets**: To be prudent, we should keep all tweets from a single user in the same fold.

 - **Target balance**: Since the target is well-balanced we don't need to ensure that all folds have the same ratio of labels. Also, the even balance means accuracy is a fine metric.

 - **Time consideration**: Since there is no date-time for the tweet, we will make no consideration for time, but in general we can only build models on data from the past, so this should be included in future cross-validation schemes. 

 - **`names` with multiple tweets**: To be prudent, we should keep all tweets from a single user in the same fold.

 - **Target balance**: Since the target is well-balanced we don't need to ensure that all folds have the same ratio of labels. Also, the even balance means accuracy is a fine metric.


```python
def accuracy(y_true, y_pred): 
    """Returns the ratio correctly classified labels, with 1 being perfect"""
    return (y_true == y_pred).mean()

assert accuracy(np.array(['male', 'female', 'female', 'brand']), np.array(['male', 'female', 'male', 'brand'])) == .75
```

```python
from sklearn.model_selection import GroupKFold

def add_cv_folds(td): 
    """Returnd td with added `fold` column"""
    
    np.random.seed(0) # for reproducibility
    td = td.loc[np.random.permutation(td.index)] # to shuffle rows
    td = td.reset_index(drop=True)
    td['fold'] = 0
    kf = GroupKFold(n_splits=5)
    for fold, (trn_id, tst_id) in enumerate(kf.split(td, groups=td.name)): 
        print("Asserting there is no overlap in `name` between train and test sets")
        assert len(set(td.loc[trn_id].name).intersection(set(td.loc[tst_id].name))) == 0  
        td.loc[tst_id, 'fold'] = fold
    return td
td = preprocess_twitter_data(pd.read_csv(data/'twitter_data.csv'))
td = add_cv_folds(td)
```

```python
from multiprocessing import Pool, cpu_count
import time
import Levenshtein

def closest_match(username, names):
    """
    Finds the name in a list that is closest to a given username.
    """
    min_distance = float('inf')
    closest_name = None
    for name in names:
        distance = Levenshtein.distance(username, name)
        if distance < min_distance:
            min_distance = distance
            closest_name = name

    return closest_name, min_distance

def pool_func(function, input_list: list, verbose=False, n_cpu=99):
    """Uses the Pool function from the package 'multiprocessing'
    to run `function` over the list `input_list`.  The `function`
    should only take """

    n_cpu = min(n_cpu, cpu_count())
    if verbose:
        print('#############################################')
        print('Pooling function: ')
        if hasattr(function, '__name__'):
            print(function.__name__)
        print(f'{n_cpu} of {cpu_count()} cpus used')
        print('Number of function calls: ', len(input_list))

    start = time.time()
    pool = Pool(n_cpu)
    res = pool.map(function, input_list)
    pool.close()
    pool.join()

    if verbose:
        print('Time taken:',
              round((time.time() - start) / 60, 2),
              'minutes')
    return res if res else []
```

```python
def get_dataloader(df, tokenizer, batch_size=16): 
    user_descriptions = df.user_description.fillna('NaN') # list of user descriptions
    tweet_texts = df.tweet_text.fillna('NaN') # list of tweet texts
    labels = df.gender.map(label_to_num).tolist() # list of labels
    encoded_dict = tokenizer.batch_encode_plus(list(zip(user_descriptions, tweet_texts)), padding=True, truncation=True, max_length=128)
    input_ids = torch.tensor(encoded_dict['input_ids'])
    attention_masks = torch.tensor(encoded_dict['attention_mask'])
    labels = torch.tensor(labels)
    dataset = TensorDataset(input_ids, attention_masks, labels)
    return DataLoader(dataset, batch_size=batch_size)
```

```python
class BertTuner: 
    def __init__(self, epochs=3, lr=2e-5): 
        self.epochs = epochs
        self.lr = lr
        self.tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
        
    def fit(self, X, y): 
        
        model = BertForSequenceClassification.from_pretrained('bert-base-uncased', num_labels=3).to(device)
        
        dataloader = get_dataloader(X, self.tokenizer)
        optimizer = torch.optim.Adam(model.parameters(), lr=self.lr)
        for epoch in trange(self.epochs):
            for batch in dataloader:
                model.train()
                batch = tuple(t.to(device) for t in batch)
                inputs = {'input_ids': batch[0], 'attention_mask': batch[1], 'labels': batch[2]}
                outputs = model(**inputs)
                loss = outputs[0]
                loss.backward()
                optimizer.step()
                optimizer.zero_grad()
        self.model = model
        
    def predict(self, X): 
        model = self.model
        model.eval()
        dataloader = get_dataloader(X, self.tokenizer)
        preds = []
        for batch in tqdm(dataloader):
            batch = tuple(t.to(device) for t in batch)
            inputs = {'input_ids': batch[0], 'attention_mask': batch[1], 'labels': batch[2]}
            with torch.no_grad():
                outputs = model(**inputs)
            logits = outputs[1]
            preds.append(logits.to('cpu'))
        return [num_to_label[x] for x in torch.concat(preds).tolist()]        
    
    def predict_proba(self, X): 
        model = self.model
        model.eval()
        dataloader = get_dataloader(X, self.tokenizer)
        preds = []
        for batch in tqdm(dataloader):
            batch = tuple(t.to(device) for t in batch)
            inputs = {'input_ids': batch[0], 'attention_mask': batch[1], 'labels': batch[2]}
            with torch.no_grad():
                outputs = model(**inputs)
            logits = outputs[1]
            preds.append(logits.cpu().numpy())
        return np.vstack(preds)
```

## gpt2 embeddings

 * note about using mean(dim=1) vs [:, -1, :]: I don't know whats better or what I'm missing but chatgpt says 
  > For the specific task of classifying a user as 'male', 'female', or 'brand' from a tweet, I would recommend using the hidden state of the last token, which can be obtained using [:, -1, :]. This is because the hidden state of the last token typically contains the most relevant information for the task of generating the next token or word in the sequence, and in this case, the task is to classify the user.

    > Using the mean of the hidden states across the sequence dimension, which can be obtained using .mean(dim=1), may be more suitable for tasks like sentiment analysis or text classification, where the model needs to generate a fixed-length representation of the input sequence.

```python
def blend_mean_with_weights(preds: tuple, weights: tuple, num_to_label: callable=None): 
    assert len(preds) == len(weights)
    res = preds[0] * weights[0]
    for p, w in zip(preds[1:], weights[1:]): 
        res += p * w
    res = res / len(preds)
    if num_to_label: return proba_to_preds(res, num_to_label)
    else: return res

def blend_mean_with_maximum(preds: tuple, num_to_label: callable=None): 
    res = np.maximum(*preds)
    if num_to_label: return proba_to_preds(res, num_to_label)
    else: return res
```

```python
class BertEmbTuner: 
    def __init__(self, epochs=3, lr=2e-5, get_dataloader_func:callable=get_dataloader): 
        self.epochs = epochs
        self.lr = lr
        self.tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')
        self.get_dataloader_func = get_dataloader_func
        
    def fit(self, X, y): 
        
        model = BertForSequenceClassification.from_pretrained(
            'bert-base-uncased', output_hidden_states=True, num_labels=3).to(device)
        
        dataloader = self.get_dataloader_func(X, self.tokenizer)
        optimizer = torch.optim.Adam(model.parameters(), lr=self.lr)
        for epoch in trange(self.epochs):
            for batch in dataloader:
                model.train()
                batch = tuple(t.to(device) for t in batch)
                inputs = {'input_ids': batch[0], 'attention_mask': batch[1], 'labels': batch[2]}
                outputs = model(**inputs)
                loss = outputs[0]
                loss.backward()
                optimizer.step()
                optimizer.zero_grad()
        self.model = model
        
    def predict(self, X): 
        model = self.model
        model.eval()
        dataloader = get_dataloader(X, self.tokenizer)
        preds, embs = [], []
        for batch in tqdm(dataloader):
            batch = tuple(t.to(device) for t in batch)
            inputs = {'input_ids': batch[0], 'attention_mask': batch[1], 'labels': batch[2]}
            with torch.no_grad():
                outputs = model(**inputs)
            logits = outputs[1]
            preds.append(logits.to('cpu').numpy())
            embeddings = outputs.hidden_states[-1].mean(dim=1).cpu().numpy()
            embs.append(embeddings)
        return {'probas': np.vstack(preds), 'embs': np.vstack(embs)}
#         return {'probas': preds, 'embs': embs}    
```

```python
def add_preds_to_df(d, df, mask, prefix=''): 
    proba_cols = [f'{prefix}proba_{x}' for x in range(d['probas'].shape[1])]
    emb_cols = [f'{prefix}emb_{x}' for x in range(d['embs'].shape[1])]
    df.loc[mask, proba_cols] = d['probas']
    df.loc[mask, emb_cols] = d['embs']
    
def accumulate_preds_to_df(d, df, prefix=''): 
    proba_cols = [f'{prefix}proba_{x}' for x in range(d['probas'].shape[1])]
    emb_cols = [f'{prefix}emb_{x}' for x in range(d['embs'].shape[1])]
    if all(col in df.columns for col in proba_cols):
        df.loc[:, proba_cols] += d['probas']
        df.loc[:, emb_cols] += d['embs']       
    else: 
        df.loc[:, proba_cols] = d['probas']
        df.loc[:, emb_cols] = d['embs']   
```
